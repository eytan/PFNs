```python
fig, axs = plt.subplots(1, 2, figsize=(8, 3))

df = pd.read_csv("/home/lily_l/private_multitask_pfn/benchmark_time_samples.csv")
n_samples_per_task = df["n_samples_per_task"]
mtgp = df["mtgp"]
scaml = df["scaml"]
pfn = df["pfn"]

axs[0].plot(n_samples_per_task, mtgp, label="Joint (ICM)")
axs[0].plot(n_samples_per_task, scaml, label="Ensemble (ScaML)")
axs[0].plot(n_samples_per_task, pfn, label="MTPFN")

axs[0].set_xlabel("Number of samples per task")
axs[0].set_ylabel("Time (s)")
axs[0].set_ylim(-0.5, 13)
axs[0].set_title("5 Tasks")

df = pd.read_csv("/home/lily_l/private_multitask_pfn/benchmark_time_task.csv")
n_tasks = df["n_tasks"]
mtgp = df["mtgp"]
scaml = df["scaml"]
pfn = df["pfn"]


axs[1].plot(n_tasks, mtgp, label="Joint (ICM)")
axs[1].plot(n_tasks, scaml, label="Ensemble (ScaML)")
axs[1].plot(n_tasks, pfn, label="MTPFN")

axs[1].set_title("50 Samples per Task")
axs[1].set_xlabel("Number of tasks")
axs[1].set_ylabel("Time (s)")
axs[1].set_xscale("log")

axs[1].set_ylim(-0.5, 13)

# for axis set grid
for ax in axs:
    ax.grid(True, which='major', linestyle='--', linewidth=0.5)

plt.legend()
plt.savefig("benchmark_time.pdf", bbox_inches='tight')


# plt.legend()
# plt.show()
```
